```python
# 图片分割与合并
import cv2
import numpy as np

img = np.zeros((200,300,3),np.uint8)
b,g,r = cv2.split(img)
b[10:50,10:50] = 255
img2 = cv2.merge((b,g,r))
cv2.imshow ('img',np.hstack((img,img2)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 绘制各种图行
import cv2
import numpy as np
img = np.zeros((480,640,3),np.uint8)
# line(img,pt1,pt2,color,thickness,lineType,shift) 缩放比0,4，8
cv2.line(img,(10,20),(10,300),(0,0,255),5,4)
cv2.line(img,(10,20),(300,400),(0,0,255),5,16)


cv2.imshow('drow',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
